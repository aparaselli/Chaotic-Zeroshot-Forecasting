```python
import matplotlib.pyplot as plt
import pandas as pd
import json
import os
```

```python
with open('C:\\Users\\hirot\\Documents\\Chaotic-Zeroshot-Forecasting\\scripts\\training\\output\\run-6\\checkpoint-200000\\trainer_state.json') as js:
    d = json.load(js)

with open('C:\\Users\\hirot\\Documents\\Chaotic-Zeroshot-Forecasting\\scripts\\training\\output\\run-8\\checkpoint-200000\\trainer_state.json') as js:
    e = json.load(js)

with open('C:\\Users\\hirot\\Documents\\Chaotic-Zeroshot-Forecasting\\scripts\\training\\output\\mixed-data\\checkpoint-200000\\trainer_state.json') as js:
    f = json.load(js)

with open('C:\\Users\\hirot\\Documents\\Chaotic-Zeroshot-Forecasting\\scripts\\training\\output\\mixed-data-mini\\checkpoint-200000\\trainer_state.json') as js:
    g = json.load(js)
```

```python
df = pd.DataFrame(d['log_history'])
df['epoch'] = df['epoch'] * 200_000

ef = pd.DataFrame(e['log_history'])
ef['epoch'] = ef['epoch'] * 200_000

ff = pd.DataFrame(f['log_history'])
ff['epoch'] = ff['epoch'] * 200_000

gf = pd.DataFrame(g['log_history'])
gf['epoch'] = gf['epoch'] * 200_000
```

```python
plt.plot(df['epoch'], df['loss'], label = 'Experiment #2')
plt.plot(ef['epoch'], ef['loss'], label = 'Experiment #3')
plt.plot(ff['epoch'], ff['loss'], label = 'Experiment #4')
plt.plot(gf['epoch'], gf['loss'], label = 'Experiment #5')


plt.title('SMAPE Loss vs Epochs', fontdict={'size': 14})
plt.xlabel('Epochs', fontdict={'size': 14})
plt.ylabel('SMAPEL Loss', fontdict={'size': 14})
plt.tick_params(axis='both', labelsize=12)
plt.xticks(rotation=25);  # Tilt x-axis labels by 25 degrees
plt.legend();
```

```python
plt.plot(df['epoch'], df['grad_norm'], label = 'Experiment #1')
plt.plot(ef['epoch'], ef['grad_norm'], label = 'Experiment #2')
plt.plot(ff['epoch'], ff['grad_norm'], label = 'Experiment #3')
plt.plot(gf['epoch'], gf['grad_norm'], label = 'Experiment #4')


plt.title('Grad Norm vs Epochs', fontdict={'size': 14})
plt.xlabel('Epochs', fontdict={'size': 14})
plt.ylabel('Grad Norm', fontdict={'size': 14})
plt.tick_params(axis='both', labelsize=12)
plt.xticks(rotation=25);  # Tilt x-axis labels by 25 degrees
plt.legend();
```
