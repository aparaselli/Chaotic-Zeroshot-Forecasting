```python
from dysts.systems import make_trajectory_ensemble
import json

all_out = make_trajectory_ensemble(324800, resample=False, pts_per_period=100)
rows = []

# Process each attractor
for attractor_name, values in all_out.items():
    # Create a dictionary for this attractor's row
    
    # Add columns for each variable (x, y, z) at each time step
    for var_idx, var_name in enumerate([f'x_{I}' for I in range(values.shape[1])]):  # Iterate through variables
        row = {'attractor': f'{attractor_name}_{var_name}'}
        row['values'] = values[:,var_idx]
        rows.append(row)

# Create DataFrame from the list of rows
df = pd.DataFrame(rows)
def prepare_for_json(val):
    if isinstance(val, np.ndarray):
        return val.tolist()  # Convert NumPy array to Python list
    return val

# First convert arrays to lists, then to JSON strings
df['values'] = df['values'].apply(prepare_for_json).apply(json.dumps)
df.to_csv('attractor_data.csv', index=False)
df
```
