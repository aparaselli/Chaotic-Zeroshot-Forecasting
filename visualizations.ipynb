## Find ideal timestep

```python
import pandas as pd 
import ast
import numpy as np
import matplotlib.pyplot as plt
```

```python
attractors = pd.read_csv('Data/attractor_data.csv')
attractors.head(5)
```

```python
attractors['values'] = attractors['values'].apply(lambda x: ast.literal_eval(x))
```

```python
def sample_attractor(attractor_list,sample_length,step_sz='Random',start_pt='Random'):
    if step_sz == 'Random':
        step_sz = np.random.randint(80,200)
        print(f'Step size is {step_sz}')
    if start_pt == 'Random':
        start_pt = np.random.randint(0,len(attractor_list)-(sample_length*step_sz))
    return attractor_list[start_pt:start_pt+(sample_length*step_sz):step_sz]

sample_attractor(attractors['values'].iloc[0],20)
```

```python
plt.plot(sample_attractor(attractors['values'].iloc[81],576))
```
